```python
"""df=n-1, sem=s/sqrt(n), t=(xbar-u)/sem, 
    ci=xbar+-moe, moe=tcrit*sem, cohend=(xbar-u)/s, 
    r^2=t^2/(t^2+df)
"""
xs1 = 14.05
xs2 = 22.02
ns1 = 24
ss = stsample["Squared Diff"].sum()
v = ss/(ns1 - 1)
s = mt.sqrt(v)
df = (ns1 -1)
md = xs1 - xs2
tcrit = 1.714


#sp2 = (ss1 + ss2)/float(df)

#sem = mt.sqrt((sp2/ns1) + (sp2/ns2))
sem = s/mt.sqrt(ns1)
#t = (xs1 - xs2 - 3)/ sem
t = md/sem
t, df
```

```python
cod = md/s
rsq = (t * t)/((t * t) + df)

cod, rsq
```
