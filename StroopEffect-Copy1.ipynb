# Stroop Effect

1.
    The independent variable is the time it takes to complete each list of words. The dependent variable is the state of the words, congruent or incongruent.

2.
    My hypothesis for this test is we will reject H<sub>0</sub>. H<sub>0</sub> being that incongruence will have no effect on the time it takes to complete the test. I will use a one tailed Z-test for this data. The data is numerical and measures an effect on subjects. This will compare the mean of one set of data against the mean of another from the same subjects.
    
3.
    

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = "stroopdata.csv"

stsample = pd.read_csv(filename)
stsample
```

```python
index=["Mean", "Median", "Std Dev"]
constats = pd.Series([stsample["Congruent"].mean(), stsample["Congruent"].median(), stsample["Congruent"].std()], index=index)
incostats = pd.Series([stsample["Incongruent"].mean(), stsample["Incongruent"].median(), stsample["Incongruent"].std()], index=index)
statsdf = pd.DataFrame({"Congruent": constats, "Incongruent": incostats})
statsdf
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(14,6), sharey=True)
ax1.bar(len(stsample), stsample[["Congruent"]], color="b")
ax2.bar(len(stsample), stsample[["Incongruent"]], color="r")

plt.show()
```

```python
conpx = stsample[["Congruent"]].sort_values(by="Congruent").plot(kind='bar', figsize=(7, 7), color=['b'])
conpx.set_title("Congruent Word Times ", fontsize='18')
conpx.set_xlabel('Fare Price', fontsize='14')
conpx.set_ylabel('Survival %', fontsize='14')
plt.show()
```
